```python
import requests
from bs4 import BeautifulSoup
import pandas
import sqlite3
import logging
import re
from urllib.parse import quote_plus
import random

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

try:
    knjige_za_pronaci_df = pandas.read_csv('knjige.csv')
    logging.info("Uspesno ucitan fajl")
except FileNotFoundError:
    logging.info("Fajl nije pronadjen")
    knjige_za_pronaci_df = pandas.DataFrame()

#display(knjige_za_pronaci_df)

sve_knjige = []

for index, row in knjige_za_pronaci_df.iterrows():
    naziv_knjige = row['naslov']
    ime_autora = row['autor']
    logging.info(naziv_knjige + " " + ime_autora)

    knjiga_info = {}

    try:
        #api_url = f"https://openlibrary.org/search?q={quote_plus(naziv_knjige)}" //stranica
        api_url = f"https://openlibrary.org/search.json?title={quote_plus(naziv_knjige)}&author={quote_plus(ime_autora)}" #json
        logging.info(api_url)
        #print(api_url)
    
        response = requests.get(api_url)
        response.raise_for_status()
        data = response.json()
        #print(data)
    
        if 'docs' in data and len(data['docs']) > 0:
            result = data['docs'][0]
            #display(result)
    
            knjiga_info['godina_izdavanja'] = result.get('first_publish_year')
            knjiga_info['broj_edicija'] = result.get('edition_count')
            knjiga_info['id_autora'] = result.get('author_key')
            knjiga_info['ime_autora'] = result.get('author_name')
            knjiga_info['naziv_knjige'] = result.get('title')
        else:
            knjiga_info['ime_autora'] = ime_autora
            knjiga_info['naziv_knjige'] = naziv_knjige
    except Exception as e:
        logging.error(e)
        knjiga_info['ime_autora'] = ime_autora
        knjiga_info['naziv_knjige'] = naziv_knjige

    try:
        autor_za_pretragu = knjiga_info['ime_autora'][0]
        #print(autor_za_pretragu)

        wiki_url_tmp = f"https://en.wikipedia.org/wiki/{quote_plus(autor_za_pretragu)}"
        logging.info(wiki_url_tmp)
        wiki_url = wiki_url_tmp.replace('+', '_')
        logging.info("wiki_url " + wiki_url)

        response = requests.get(wiki_url)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'lxml')
        infobox = soup.find('table', class_='infobox vcard')
        autor_detalji = {}
        if infobox:
            for infobox_row in infobox.find_all('tr'):
                header = infobox_row.find('th')
                data_cell = infobox_row.find('td')
                if header and data_cell:
                    header_text = header.text.strip()
                    if 'Born' in header_text:
                        autor_detalji['born'] = data_cell.text.strip()
        knjiga_info['autor_detalji'] = autor_detalji
        knjiga_info['broj_strana'] = random.randint(300, 1000)
        #print("knjiga info_______")
        #display(knjiga_info)
    except Exception as e:
        logging.error(e)
    sve_knjige.append(knjiga_info)

#print("sve_knjige")
#display(sve_knjige)

autori = {} #koristimo skup strukturu jer ne zelimo da nam se ponavljaju autori, jer skup nema duplikata

for knjiga in sve_knjige:
    ime = knjiga.get('ime_autora')[0]
    if ime and ime not in autori:
        detalji = knjiga.get('autor_detalji', {})
        rodjenje_info = detalji.get('born', '')
        match = re.search(r'\b(\d{4})\b', rodjenje_info)
        godina_rodjenja = int(match.group(1)) if match else None
        id = knjiga.get('id_autora')[0]

        autori[ime] = {
            'ime': ime,
            'godina_rodjenja': godina_rodjenja,
            'id': id
        }

autori_df = pandas.DataFrame(list(autori.values()))
autori_df['godina_rodjenja'] = pandas.to_numeric(autori_df['godina_rodjenja']).astype('Int64')
display(autori_df)

knjige = [] #koristimo listu jer ovde smeju da postoje duplikati

for knjiga in sve_knjige:
    jedna_knjiga = {
        'naslov': knjiga.get('naziv_knjige'),
        'autor': knjiga.get('ime_autora')[0],
        'godina_izdavanja': knjiga.get('godina_izdavanja'),
        'broj_strana': knjiga.get('broj_strana')
    }
    knjige.append(jedna_knjiga)

knjige_df = pandas.DataFrame(knjige)
knjige_df['godina_izdavanja'] = pandas.to_numeric(knjige_df['godina_izdavanja']).astype('Int64')
knjige_df['broj_strana'] = pandas.to_numeric(knjige_df['broj_strana']).astype('Int64')
display(knjige_df)

db_file = 'knjige_db.db'
conn = sqlite3.connect(db_file)

autori_df.to_sql('autori', conn, if_exists='replace')
knjige_df.to_sql('knjige', conn, if_exists='replace')

conn.close()

conn = sqlite3.connect(db_file)

query = """
select
    k.naslov,
    k.godina_izdavanja,
    a.ime,
    a.godina_rodjenja,
    a.id
from
    knjige k
join 
    autori a on k.autor = a.ime
where
    a.godina_rodjenja < 1930;
"""

results_df = pandas.read_sql_query(query, conn)
conn.close()

display(results_df)
```

```python
import pandas
import sqlite3
import matplotlib.pyplot as plt
import seaborn

db_file = 'knjige_db.db'
conn = sqlite3.connect(db_file)

query = """
select
    k.naslov as naslov_knjige,
    k.godina_izdavanja,
    k.broj_strana,
    a.ime as ime_autora,
    a.godina_rodjenja,
    a.id as id_autora
from
    knjige k
join 
    autori a on k.autor = a.ime
"""

df = pandas.read_sql_query(query, conn)
conn.close()

#df['godina_rodjenja'] = pandas.to_numeric(df['godina_rodjenja']).astype('Int64')

#display(df)
#df.info()

df.dropna(subset=['godina_rodjenja'], inplace=True) #inplace=True menja postojeci df i iz njega izbacuje red koji treba
#display(df)

df['godina_izdavanja'] = df['godina_izdavanja'].astype(int)
df['broj_strana'] = df['broj_strana'].astype(int)
df['godina_rodjenja'] = df['godina_rodjenja'].astype(int)

display(df)
df.info()

plt.figure(figsize=(10, 6))
seaborn.histplot(data=df, x='broj_strana')

plt.title('Distribucija broja stranica knjiga')
plt.xlabel('Broj stranica')
plt.ylabel('Broj knjiga')
plt.show()

display(df['broj_strana'].describe())

plt.figure(figsize=(10, 6))
seaborn.scatterplot(data=df, x='godina_izdavanja', y='broj_strana', hue='ime_autora') #hue grupise podatke na osnovu kolone u ovom slucaju ime _autora i svakom autoru dodeljuje razlicitu boju

plt.title('Odnos godine izdavanja i broja strana')
plt.xlabel('Godina izdavanja')
plt.ylabel('Broj stranica')
plt.show()
```
