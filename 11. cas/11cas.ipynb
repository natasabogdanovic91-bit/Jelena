```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

URL = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"

try: #kad negde ocekujes error pa stavis u try block
    response = requests.get(URL)
    response.raise_for_status()
    print(f"Uspesan request status kod: {response.status_code}")
except requests.exceptions.RequestException as e:
    print(f"Greska prilikom requesta: {e}")

soup = BeautifulSoup(response.content,'html.parser')

#print(soup.prettify())

table = soup.find('table', class_='wikitable')
if table is None: 
    raise RuntimeError("Nije pronadjena tabela") #kad funkcija ne generise error kakav nam treba, ali mi hocemo da imamo error u odredjenoj situaciji

#print(table)

headers = [th.text.strip() for th in table.find_all('th')] #strip sklanja prazna mesta, tabulatore i nove redove
print(headers)

rows = [] #tr tagovi

for row in table.find('tbody').find_all('tr')[1:]:
    cells = row.find_all('td')
    if len(cells) > 1:
        row_data = [cell.text.strip() for cell in cells]
        rows.append(row_data)
#print(rows)

df = pd.DataFrame(rows, columns=headers)
display(df.head())
```
