```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("senzori_log.csv", parse_dates=["vreme"])

# Filtriraj po senzoru
senzor_id = 1
df_senzor = df[df["senzor_id"] == senzor_id]

plt.plot(df_senzor["vreme"], df_senzor["temperatura"], label="Temperatura")
plt.plot(df_senzor["vreme"], df_senzor["vlaga"], label="Vlaga")
plt.xlabel("Vreme")
plt.ylabel("Vrednost")
plt.title(f"Senzor {senzor_id}: temperatura i vlaga")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#

senzori = df["senzor_id"].unique()
fig, axs = plt.subplots(len(senzori), 1, figsize=(8, 4*len(senzori)))

for i, sid in enumerate(senzori):
    podaci = df[df["senzor_id"] == sid]
    axs[i].plot(podaci["vreme"], podaci["temperatura"], label="Temperatura")
    axs[i].plot(podaci["vreme"], podaci["vlaga"], label="Vlaga")
    axs[i].set_title(f"Senzor {sid}")
    axs[i].legend()
    axs[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("prodaja.csv")
df["vrednost"] = df["kolicina"] * df["cena"]

# Grupisanje po datumu
po_danu = df.groupby("datum")["vrednost"].sum()

plt.plot(po_danu.index, po_danu.values, marker='o')
plt.title("Ukupna dnevna prodaja")
plt.xlabel("Datum")
plt.ylabel("Vrednost ()")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Grupisanje po proizvodu
po_proizvodu = df.groupby("proizvod")["vrednost"].sum()

plt.bar(po_proizvodu.index, po_proizvodu.values, color='orange')
plt.title("Ukupna prodaja po proizvodima")
plt.xlabel("Proizvod")
plt.ylabel("Vrednost ()")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("prodaja_slozeno.csv")
df["vrednost"] = df["kolicina"] * df["cena"]

# Ukupna prodaja po regionu
po_regionu = df.groupby("region")["vrednost"].sum()
print("Ukupna prodaja po regionu:\n", po_regionu)

# Prosena cena po kategoriji
prosecna_cena = df.groupby("kategorija")["cena"].mean()
print("\nProsena cena po kategoriji:\n", prosecna_cena)

# Vizualizacija: ukupna vrednost po danima i regionima
po_danu_regionu = df.groupby(["datum", "region"])["vrednost"].sum().unstack()

po_danu_regionu.plot(figsize=(12,6), marker='o')
plt.title("Vrednost prodaje po regionima kroz vreme")
plt.xlabel("Datum")
plt.ylabel("Vrednost ()")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("web_logs.csv", parse_dates=["timestamp"])

# Broj status kodova
print(df["status_code"].value_counts())

# Broj zahteva po korisniku
zahtevi_po_korisniku = df["user_id"].value_counts()
zahtevi_po_korisniku.plot(kind='bar', title="Broj zahteva po korisniku")
plt.ylabel("Broj zahteva")
plt.tight_layout()
plt.show()

# Broj status kodova po satu
df["sat"] = df["timestamp"].dt.floor("h")
status_po_satu = df.groupby(["sat", "status_code"]).size().unstack(fill_value=0)

status_po_satu.plot(figsize=(12,6))
plt.title("Status kodovi kroz vreme")
plt.xlabel("Vreme")
plt.ylabel("Broj zahteva")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("iot_metrics.csv", parse_dates=["timestamp"])

# Prosene vrednosti po ureaju
prosek_po_uredjaju = df.groupby("device_id")[["temperature", "humidity"]].mean()
print("Prosek po ureaju:\n", prosek_po_uredjaju)

# Vizualizacija po ureaju
uredjaji = df["device_id"].unique()

fig, axs = plt.subplots(len(uredjaji), 2, figsize=(14, 4 * len(uredjaji)))

for i, dev in enumerate(uredjaji):
    data = df[df["device_id"] == dev]
    axs[i][0].plot(data["timestamp"], data["temperature"], label="Temp", color='red')
    axs[i][0].set_title(f"{dev} - Temperatura")
    
    axs[i][1].plot(data["timestamp"], data["humidity"], label="Vlaga", color='blue')
    axs[i][1].set_title(f"{dev} - Vlanost")

for ax_row in axs:
    for ax in ax_row:
        ax.set_xlabel("Vreme")
        ax.set_ylabel("Vrednost")
        ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Uitavanje i spajanje
dfs = []
for mesec in ["januar", "februar", "mart"]:
    df = pd.read_csv(f"{mesec}.csv")
    dfs.append(df)

df_sve = pd.concat(dfs)

# Dodavanje kolone za prihod
df_sve["prihod"] = df_sve["kolicina"] * df_sve["cena"]

# Agregacija
ukupan_prihod = df_sve.groupby(["region", "proizvod"])["prihod"].sum().reset_index()

# Top 5 proizvoda po prihodu
top5 = ukupan_prihod.groupby("proizvod")["prihod"].sum().nlargest(5)
print(top5)

# Vizualizacija po regionima
df_region = df_sve.groupby("region")["prihod"].sum()
df_region.plot(kind="bar", title="Prihod po regionima")
plt.ylabel("Prihod")
plt.show()
```
